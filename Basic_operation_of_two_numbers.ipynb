```python
from tkinter import *
import tkinter as tk

root= tk.Tk()

root.resizable(0, 0)

root.title("Calcualtor")

canvas1 = tk.Canvas(root, width = 600, height = 300,  relief = 'raised', bg="white")
canvas1.pack()

label1 = tk.Label(root, text='Calculations of two Numbers')
label1.config(font=('helvetica', 16),bg="white")
canvas1.create_window(300, 30, window=label1)

label2 = tk.Label(root, text='Enter the First number:')
label2.config(font=('helvetica', 12),bg="white")
canvas1.create_window(120, 90, window=label2)

entry1 = tk.Entry(root, font = ('helvetica', 12, 'bold'), borderwidth=2)
canvas1.create_window(380, 90, window=entry1)

label3 = tk.Label(root, text='Enter the Second number:')
label3.config(font=('helvetica', 12),bg="white")
canvas1.create_window(130, 130, window=label3)

entry2 = tk.Entry (root, font = ('helvetica', 12, 'bold'), borderwidth=2) 
canvas1.create_window(380, 130, window=entry2)

result = DoubleVar()

entry3 = tk.Entry (root, font = ('helvetica', 12, 'bold'), borderwidth=2,state='readonly', width=10, textvariable=result) 
canvas1.create_window(370, 250, window=entry3)


def getSum ():
    
    x1 = entry1.get()
    x2 = entry2.get()
    
    add = float(x1)+float(x2)
    result.set(str(add))
    
    label4 = tk.Label(root, text= 'Sum of ' + x1 +' & '+ x2 +' is:',font=('helvetica', 14),bg="white")
    canvas1.create_window(200, 250, window=label4)
    
def getSub ():
    
    x1 = entry1.get()
    x2 = entry2.get()
    
    sub = float(x1)-float(x2)
    result.set(str(sub))
    
    label4 = tk.Label(root, text= 'Sub  of ' + x1 +' & '+ x2 +' is:',font=('helvetica', 14),bg="white")
    canvas1.create_window(200, 250, window=label4)
     
def getMul ():
    
    x1 = entry1.get()
    x2 = entry2.get()
    
    mul = float(x1)*float(x2)
    result.set(str(mul))
    
    label4 = tk.Label(root, text= 'Mul of ' + x1 +' & '+ x2 +' is:',font=('helvetica', 14),bg="white")
    canvas1.create_window(200, 250, window=label4)
    
def getDiv ():
    
    x1 = entry1.get()
    x2 = entry2.get()
    
    div = float(x1)/float(x2)
    result.set(str(div))
    
    label4 = tk.Label(root, text= 'Div of ' + x1 +' & '+ x2 +' is:',font=('helvetica', 14),bg="white")
    canvas1.create_window(200, 250, window=label4)
    
def getMod ():
    
    x1 = entry1.get()
    x2 = entry2.get()
    
    div = float(x1)%float(x2)
    result.set(str(div))
    
    label4 = tk.Label(root, text= 'Mod of ' + x1 +' & '+ x2 +' is:',font=('helvetica', 14),bg="white")
    canvas1.create_window(200, 250, window=label4)
    
button1 = tk.Button(text='Addition', command=getSum, bg='black', fg='white', font=('helvetica', 12, 'bold'))
canvas1.create_window(75, 200, window=button1)

button2 = tk.Button(text='Substraction', command=getSub, bg='black', fg='white', font=('helvetica', 12, 'bold'))
canvas1.create_window(185, 200, window=button2)

button3 = tk.Button(text='Multiplication', command=getMul, bg='black', fg='white', font=('helvetica', 12, 'bold'))
canvas1.create_window(315, 200, window=button3)

button4 = tk.Button(text='Division', command=getDiv, bg='black', fg='white', font=('helvetica', 12, 'bold'))
canvas1.create_window(425, 200, window=button4)

button5 = tk.Button(text='Modulus', command=getMod, bg='black', fg='white', font=('helvetica', 12, 'bold'))
canvas1.create_window(520, 200, window=button5)

root.mainloop()
```
